```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
```

```python
n=50 #clientes 
clientes = [x for x in range (1,n+1)]
nodos = [0]+clientes 

rnd = np.random
rnd.seed(1)
```

```python
#Crear coordenadas
rnd.seed(1)
loc_x = rnd.rand(len(nodos))*100
loc_y = rnd.rand(len(nodos))*100
```

```python
plt.figure(figsize=(19 ,9))
plt.scatter(loc_x,loc_y,color="green")


plt.plot(loc_x[0],loc_y[0],color='red',marker='s')
plt.annotate('DC',(loc_x[0]-2,loc_y[0]+2))

plt.title("Gr√°fico de Nodos | VRP")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

plt.show()
```
